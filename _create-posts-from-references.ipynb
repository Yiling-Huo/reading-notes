````python
import os, re
from unidecode import unidecode

# os.chdir(os.path.dirname(os.path.abspath(__file__)))

latex_replacements = {
    r'{\\\'{a}}': 'á', r'{\\\'{e}}': 'é', r'{\\\'{i}}': 'í', r'{\\\'{o}}': 'ó', r'{\\\'{u}}': 'ú', r'{\\\'{c}}': 'ć', r'{\\c{c}}': 'ć', r'{\\\'\\i}': 'í',
    r'{\\\"{a}}': 'ä', r'{\\\"{o}}': 'ö', r'{\\\"{u}}': 'ü', r'{\\\"{i}}': 'ï', r'{\\\"{e}}': 'ë',
    r'{\\~{n}}': 'ñ', r'{\\~{a}}': 'ã', r'{\\~{o}}': 'õ', r'{\\`{a}}': 'à', r'{\\`{e}}': 'è',
    r'{\\`{i}}': 'ì', r'{\\`{o}}': 'ò', r'{\\`{u}}': 'ù', r'{\\{o}}': 'ø',
    r'{\\\'a}': 'á', r'{\\\'e}': 'é', r'{\\\'i}': 'í', r'{\\\'o}': 'ó', r'{\\\'u}': 'ú', r'{\\\'c}': 'ć',
    r'{\\\"a}': 'ä', r'{\\\"o}': 'ö', r'{\\\"u}': 'ü', r'{\\\"i}': 'ï', r'{\\\"e}': 'ë',
    r'{\\~n}': 'ñ', r'{\\~a}': 'ã', r'{\\~o}': 'õ', r'{\\`a}': 'à', r'{\\`e}': 'è',
    r'{\\`i}': 'ì', r'{\\`o}': 'ò', r'{\\`u}': 'ù', r'{\\o}': 'ø'
}

def fix_title(title, file_path):
    # if file already exists, check if the title in the file and here is the same, if so, skip everything, if not, append '-a' to the file name. 
    if os.path.exists(file_path):
        with open(file_path,'r', encoding='utf-8') as check_file:
            filec = [line for line in check_file]
            for line in filec:
                if line[:5] == 'title':
                    if title.lower() == line[7:].lower() or title == line[7:-1].lower(): # in case the last character is line break?
                        return file_path
                    else:
                        return fix_title(title, file_path.replace('.md', '') + '-a' + '.md')
    else:
        return file_path

def create_post():
    input_list = ["F://P//refs//references-merged-clean-doi.bib"]
    for file in input_list:
        with open(file, 'r', encoding='utf-8') as inputfile:
            content = [line.replace('    ','').replace('   ','').replace('  ','').replace('\t', '').replace(' = ','').replace(' =','').replace('=','') for line in inputfile]
            inputfile.seek(0)
            content_copy = [line for line in inputfile]
            # print(content_copy[:5])
            ind = 0
            end_ind = 0
            for l in range(len(content)):
                if content[l][:5] in ['@misc','@book','@Manu']: # what is this doing here?
                    ind = l
                    print(content[l][:5])
                    title = ''
                    author_string = ''
                    author = []
                    tags = []
                    journal = ''
                    url = ''
                    year = ''
                    pass
                elif content[l][0] == '@':
                    ind = l
                    print(content[l])
                    title = ''
                    author_string = ''
                    author = []
                    tags = []
                    journal = ''
                    url = ''
                    year = ''
                elif content[l][:5] == 'title':
                    print(content[l])
                    title = content[l][5:].replace('{','').replace('},','').replace('}','').replace(': ', ' - ')
                elif content[l][:6] == 'author':
                    print(content[l])
                    author_string = content[l][6:-3]
                    # first replace irregular alphabet with its form
                    # author_string = author_string.replace('{\"{a}}', 'ä').replace('{\"{o}}', 'ö').replace('{\"{u}}', 'ü').replace('{\"{i}}', 'ï').replace('{\"{e}}', 'ë').replace("{\'{a}}", 'á').replace("{\'{e}}", 'é').replace("{\'{i}}", 'í').replace("{\'{o}}", 'ó').replace("{\'{u}}", 'ú').replace("{\'{c}}", 'ć').replace('{\~{n}}', 'ñ').replace('{\~{a}}', 'ã').replace('{\~{o}}', 'õ').replace('{\{o}}', 'ø').replace('{\`{a}}', 'à').replace('{\`{e}}', 'è').replace('{\`{i}}', 'ì').replace('{\`{o}}', 'ò').replace('{\`{u}}', 'ù').replace('{\`{u}}', 'ù')
                    # author_string = author_string.replace('{\"a}', 'ä').replace('{\"o}', 'ö').replace('{\"u}', 'ü').replace('{\"i}', 'ï').replace('{\"e}', 'ë').replace("{\'a}", 'á').replace("{\'e}", 'é').replace("{\'i}", 'í').replace("{\'o}", 'ó').replace("{\'u}", 'ú').replace("{\'c}", 'ć').replace('{\~n}', 'ñ').replace('{\~a}', 'ã').replace('{\~o}', 'õ').replace('{\o}', 'ø').replace('{\`a}', 'à').replace('{\`e}', 'è').replace('{\`i}', 'ì').replace('{\`o}', 'ò').replace('{\`u}', 'ù').replace('{\`u}', 'ù')
                    # Loop through the dictionary and replace all occurrences
                    for latex, char in latex_replacements.items():
                        author_string = re.sub(latex, char, author_string)

                    # After replacements, print the result
                    print("After replacement:", author_string)
                    # then replace the brackets
                    author_string = author_string.replace('{','').replace('}','') # is the last character the line break? .replace('},','')
                    author = author_string.split(' and ')
                    print(author)
                    try:
                        first_author_name = unidecode(author[0].split(', ')[0].lower().replace(' ','')) # use unicode version (no accents for the url)
                        print(author[0].split(', ')[1]) # ask to print the second element so that split(',') has at least two elements
                    except IndexError:
                        first_author_name = unidecode(author[0].split(' ')[-1].lower().replace(' ','')) # use unicode version (no accents for the url)
                        author_string = ' and '.join([a.split(' ')[-1]+', '+' '.join(a.split(' ')[:-1]) for a in author])
                        print(author_string)
                elif content[l][:7] == 'journal':
                    print(content[l])
                    journal = '"' + content[l][7:].replace("{{","").replace("}}","").replace('{','').replace('},','').replace(',','').replace("\\","").replace("\n","") + '"'
                elif content[l][:3] == 'doi':
                    print(content[l])
                    url = 'https://doi.org/'+content[l][3:].replace('{','').replace('},','')
                elif content[l][:8] == 'keywords':
                    tags = content[l][8:].replace('{','').replace('},','').replace('\n','').split(',')
                elif content[l][:4] == 'year':
                    print(content[l])
                    year = content[l][5:9]
                elif content[l][0] == '}':
                    end_ind = l
                    if len(title)>0 and "No Title" not in title:
                        file_path = fix_title(title, 'References/_posts/1996-01-01-'+first_author_name+'-'+str(year)+'.md')
                        try:
                            with open(file_path, 'w', encoding='utf-8') as outputfile:
                                output = '---\nlayout: post\ntitle: '+title+'\ndate: 1996-01-01 00:00\nauthor: '+author_string+'\ntags: ["'+'","'.join(tags).lower()+'"]\njournal: '+journal+'\nlink: '+url+'\nyear: '+year+'\n---'
                                output = output.replace('tags: [""]\n', '').replace('link: \n','').replace('author: \n','').replace('journal: \n','')
                                output += '\n\n```bibtex\n' + ''.join(content_copy[ind:end_ind]).replace("{{","{").replace("}}","}") + '}\n```' # add bibtex to content
                                print("begin: ", ind, "end: ", end_ind, "difference: ", end_ind - ind)
                                outputfile.write(output)
                        except OSError:
                            print("cannot create post.")
                            pass

create_post()
````

2. Clean journal commas

````python
import os

def clean_journal_commas():
    # os.chdir(os.path.dirname(os.path.abspath(__file__)))

    inputdir = 'References/_posts/'
    dataFileList = os.listdir(inputdir)

    for file in dataFileList:
        with open(inputdir + file, 'r+', encoding='utf-8') as inputfile:
            content = [line for line in inputfile]
            for i in range(len(content)):
                if content[i][:9] == 'journal: ':
                    content[i] = content[i].replace(',', '')
                    content[i] = content[i][:9] + content[i][9:].replace(':', '')
                    if len(content[i]) <= 10:
                        content.remove(content[i])
                        break
                    print(content[i])
                    print(len(content[i]))
                elif content[i][:8] == 'journal ':
                    content[i] = 'journal: '+content[i][8:]
                    print(content[i])
                    print(len(content[i]))
                # fix title commas as well
                elif content[i][:7] == 'title: ':
                    content[i] = content[i][:7] + content[i][7:].replace(':', ' -')
            inputfile.seek(0)
            inputfile.write(''.join(content))

clean_journal_commas()
````

3. Clean author periods

````python
import os

def clean_author_periods():
    # os.chdir(os.path.dirname(os.path.abspath(__file__)))

    inputdir = 'References/_posts/'
    dataFileList = os.listdir(inputdir)

    for file in dataFileList:
        with open(inputdir + file, 'r+', encoding='utf-8') as inputfile:
            content = [line for line in inputfile]
            for i in range(len(content)):
                if content[i][:8] == 'author: ':
                    content[i] = content[i].replace('.', '')
                    print(content[i])
            inputfile.seek(0)
            inputfile.write(''.join(content))

clean_author_periods()
````
